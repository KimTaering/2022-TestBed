# 리스트 분류 함수

```python
#파일명을 리스트로 분류하는 함수
def sep_data(fileName_list) :
    
    #리스트 생성
    order = list()  #순서
    date = list()   #날짜
    time = list()   #시각
    temperature = list()    #체온
    
    for i in range(len(fileName_list)) :
        fileName_list[i] = fileName_list[i].split("_")
        fileName_list[i][3] = fileName_list[i][3].split(" ")
        print(fileName_list[i])

        order.append(fileName_list[i][0])
        date.append(fileName_list[i][1])
        time.append(fileName_list[i][2])
        temperature.append(fileName_list[i][3][0])
```

```python
#날짜 리스트 date 분류
def sep_date(date_list) :
    
    #리스트 생성
    date_year = list()
    date_month = list()
    date_day = list()
    
    for i in range(len(date_list)) :

        # 연도 분류
        date_year.append(date_list[i][0:4])
#         print(date_year[i])

        # 달, 일  분류
        if (len(date_list[i])== 8) :
            date_month.append(date_list[i][4:6])
            date_day.append(date_list[i][6:])
#             print(date_month[i])
#             print(date_day[i])
        else:
            date_month.append(date_list[i][4])
            date_day.append(date_list[i][5:])
#             print(date_month[i])
#             print(date_day[i])
```

```python
#시각 리스트 time 분류
def sep_time(time_list) :
    
    #리스트 생성
    time_hour = list()
    time_min = list()

    for i in range(len(time_list)) :
        
        if (len(time_list[i]) == 2) :
            time_hour.append(time_list[i][0:1])
#             print(time_hour[i])
            time_min.append(time_list[i][1:])
#             print(time_min[i])
            
        elif (len(time_list[i]) >= 3 ) :
            time_hour.append(time_list[i][0:2])
#             print(time_hour[i])
            time_min.append(time_list[i][2:])
#             print(time_min[i])           
```

# 함수 실행

```python
sep_data(files)
```

```python
sep_date(date)
sep_time(time)
```

# 데이터프레임 전환 및 csv 파일 저장

```python
import pandas as pd
```

```python
# dataframe 전환 함수
def to_DataFrame(input_list) :
    result = pd.DataFrame(input_list) 
    return result
```

```python
# 리스트를 dataframe으로 전환
df_order = to_DataFrame(order)
df_date_year = to_DataFrame(date_year)
df_date_month = to_DataFrame(date_month)
df_date_day = to_DataFrame(date_day)
df_time_hour = to_DataFrame(time_hour)
df_time_min = to_DataFrame(time_min)
df_temp = to_DataFrame(temperature)
```

```python
# dataframe 합치기
data = pd.concat([df_order, df_date_year, df_date_month, df_date_day, df_time_hour, df_time_min, df_temp], axis = 1)
data.columns = ['order', 'year', 'month', 'day', 'hour', 'minute', 'temperature']
data
```

```python
# dataframe을 csv파일로 저장
data.to_csv("data.csv", mode = 'a',index = False)
```

```python
# dataframe을 html파일로 저장
data.to_html("data.html", index = False)
```
